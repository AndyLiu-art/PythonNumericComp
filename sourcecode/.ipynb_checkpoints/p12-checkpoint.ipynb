### 分段线性插值

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 插值点
x = np.linspace(0, 10, num=11)
y = np.cos(-x**2 / 9.0)
# 新的插值点
xnew = np.linspace(0, 10, num=1001)
# 插值结果
ynew = np.interp(xnew, x, y)
plt.plot(xnew, ynew, "-", label="linear interp")
plt.plot(x, y, "o", label="data")
plt.legend(loc="best")
plt.show()
```
